```python
import pandas as pd


url = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"

# Cargar directamente desde la URL
df = pd.read_csv(url)
df["Sex"] = df["Sex"].apply(lambda x: 1 if x == "female" else 0)

# Seleccionar columnas
X = df[["Age", "Sex", "Pclass"]]
y = df["Survived"]

# Mostrar las primeras filas
df.head()
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_Y, test_Y = train_test_split(X, y, test_size = 0.2,random_state = 15)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators= 1000, random_state=25)
rf.fit(train_X, train_Y)
rf.score(test_X, test_Y)
```

```python
import pickle

with open("modelo_random_clasificacion.pkl", "wb") as f:
    pickle.dump(rf, f)
```

```python
with open ("modelo_random_clasificacion.pkl", "rb") as f:
    rf_2 = pickle.load(f)

rf_2.score(test_X, test_Y)
```

```python
age = [4,29,65]
sex = [1,0,0]
clase = [3,1,2]

df_nuevo =pd.DataFrame({
    "Age": age,
    "Sex": sex,
    "Pclass": clase
})

rf.predict(df_nuevo)
```
